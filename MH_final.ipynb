Z-score and Logistic normalization

````python
def ZScore_norm(value_, mean_, std_) : 
  return (value_ - mean_)/std_ 

value = data['Depression(%)'] 
mean = data['Depression(%)'].mean() 
std = data['Depression(%)'].std()
normalized_value = ZScore_norm(value, mean, std)
normalized_value
````

````python
def Log_norm(value_) : 
  return 1 / (1 + np.exp(-value_))
 

value = data['Depression(%)'] 
normalized_value = Log_norm(value) 
normalized_value
````

````python
import random
def sample20(d20, percent):
    n = int(len(d20) * percent)
    random_rows = random.sample(range(len(d20)), n)
    sample = d20.iloc[random_rows]
    return sample 

data20 = sample20(data, 0.2) 
data20
````

````python
def str_sample(data, strata_col, sample_pct):
    strata_dict = {}
    for i, row in data.iterrows():
        strata = row.loc[strata_col]
        if strata in strata_dict:
            strata_dict[strata].append(row)
        else:
            strata_dict[strata] = [row]

    sample = []
    for strata, strata_data in strata_dict.items():
        strata_size = len(strata_data)
        strata_sample_size = int(strata_size * sample_pct)
        strata_sample = random.sample(strata_data, strata_sample_size)
        sample += strata_sample

    return pd.DataFrame(sample)


data30 = str_sample(data, 'Country', 0.5)
data30 
````
